```python
import pandas as pd

df = pd.read_csv("../data/cleaned.csv")

print("Before:", df.shape)

# Filter users (loose filter)
user_counts = df["user_id"].value_counts()

df = df[df["user_id"].isin(user_counts[user_counts >= 2].index)]

print("After user filter:", df.shape)

# Filter products (stronger filter)
product_counts = df["product_id"].value_counts()

df = df[df["product_id"].isin(product_counts[product_counts >= 5].index)]

print("After product filter:", df.shape)

# Sample for modeling speed
df = df.sample(50000, random_state=42)

print("After sample:", df.shape)

df.to_csv("../data/final.csv", index=False)
```
